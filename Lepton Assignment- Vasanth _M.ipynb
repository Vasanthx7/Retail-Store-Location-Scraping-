```python
#functionto store the data in csv file 
def write_dicts_to_csv_fieldnames(list_of_dicts, file_path, field_names):
    with open(file_path, 'w', newline='', encoding='utf-8') as csv_file_object:
        writer = csv.DictWriter(csv_file_object, fieldnames=field_names)
        writer.writeheader()
        for row_dict in list_of_dicts:
            writer.writerow(row_dict)

columns=['gmaps_id','store_name','address','identifier','store_description','latitude','longitude','zoom_level','creation_time','update_time','is_active']
path="Web scraping-Thangamayil store.csv"
write_dicts_to_csv_fieldnames(data,path,columns)
```
